```python
fruit_protein = {
    "Avocado":2.0,
    "Guava":2.6,
    "Blackberries":2.0,
    "Oranges":0.9,
    "Banana":1.1,
    "Apple":0.3,
    "Kiwi":1.1,
    "Pomegranate":1.7,
    "Mango":0.8,
    "Cherries":1.0
}
```

```python
import pandas as pd
s2 = pd.Series(fruit_protein, name = "Protein")
s2
```

```python
# Logical indexing to display values greater than 1
s2[s2 > 1]
```

```python
# Using the logical 'not' operator
s2[~(s2 > 1)]
```

```python
# Using the logical 'and' operator
s2[(s2 > 0.5) | (s2 <= 2)]
```

```python
# Using the logical 'and' operator
s2[(s2 > 0.5) & (s2 <= 2)]
```
